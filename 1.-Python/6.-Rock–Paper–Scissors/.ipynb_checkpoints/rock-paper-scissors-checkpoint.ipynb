#### 7. Define a function that asks the player which is the gesture he or she wants to show: 'rock', 'paper' or 'scissors'.

```python
import random                        
gestures = ["ROCK", "PAPER", "SCISSORS"]  
count = 0                             
player_score = 0                             
cpu_score = 0                      
n_rounds = 4
rounds_to_win = int(((n_rounds)/2)+1)
numeric_result = 0

while count < n_rounds:                                      
    answer = input ("Pick rock, paper or scissors: ")

    print (answer.upper())                               
    computer= random.choice(gestures)                        

    print ("Computer picks",computer)                    

    if answer.upper() == "ROCK" and computer == "ROCK":
        print ("It's a tie!")
        count += 1
        print('Your score:', player_score)
        print('Computer score:', cpu_score)
        numeric_result = 0
        print('0 for tie, 1 for lose, 2 for win:',numeric_result)
        print(count)

    elif answer.upper() == "PAPER" and computer == "PAPER":
        print ("It's a tie!")
        count += 1
        print('Your score:', player_score)
        print('Computer score:', cpu_score)
        numeric_result = 0
        print('0 for tie, 1 for lose, 2 for win:',numeric_result)
        print(count)
        
        
    elif answer.upper() == "SCISSORS" and computer == "SCISSORS":
        print ("It's a tie!")
        count += 1
        print('Your score:', player_score)
        print('Computer score:', cpu_score)
        numeric_result = 0
        print('0 for tie, 1 for lose, 2 for win:',numeric_result)
        print(count)

    elif answer.upper() == "PAPER" and computer == "ROCK":
        print ("You win!")
        count += 1
        player_score += 1
        print('Your score:', player_score)
        print('Computer score:', cpu_score)
        numeric_result = 2
        print('0 for tie, 1 for lose, 2 for win:',numeric_result)
        print(count)

    elif answer.upper() == "PAPER" and computer == "SCISSORS":
        print ("You lose!")
        count += 1
        cpu_score += 1
        print('Your score:', player_score)
        print('Computer score:', cpu_score)
        numeric_result = 1
        print('0 for tie, 1 for lose, 2 for win:',numeric_result)
        print(count)

    elif answer.upper() == "ROCK" and computer == "PAPER":
        print ("You lose!")
        count += 1
        cpu_score += 1
        print('Your score:', player_score)
        print('Computer score:', cpu_score)
        numeric_result = 1
        print('0 for tie, 1 for lose, 2 for win:',numeric_result)
        print(count)

    elif answer.upper() == "ROCK" and computer == "SCISSORS":
        print ("You win!")
        count += 1
        player_score += 1
        print('Your score:', player_score)
        print('Computer score:', cpu_score)
        numeric_result = 2
        print('0 for tie, 1 for lose, 2 for win:',numeric_result)
        print(count)

    elif answer.upper() == "SCISSORS" and computer == "ROCK":
        print ("You lose!")
        count += 1
        cpu_score += 1
        print('Your score:', player_score)
        print('Computer score:', cpu_score)
        numeric_result = 1
        print('0 for tie, 1 for lose, 2 for win:',numeric_result)
        print(count)

    elif answer.upper() == "SCISSORS" and computer == "PAPER":
        print ("You win!")
        count += 1
        player_score += 1
        print('Your score:', player_score)
        print('Computer score:', cpu_score)
        numeric_result = 2
        print('0 for tie, 1 for lose, 2 for win:',numeric_result)
        print(count)

    if (player_score - cpu_score) == rounds_to_win:
        print ("YOUR FINAL SCORE:", player_score)
        print ("COMPUTER FINAL SCORE:",cpu_score)
        print ("YOU WIN THE MATCH!")
        break

    if (cpu_score - player_score) == rounds_to_win:
        print ("YOUR FINAL SCORE:", player_score)
        print ("COMPUTER FINAL SCORE:",cpu_score)
        print ("COMPUTER WINS THE MATCH!")
        break

if player_score < cpu_score:                      
    print ("YOUR FINAL SCORE:", player_score)
    print ("COMPUTER FINAL SCORE:",cpu_score)
    print ("COMPUTER WINS THE MATCH!")

if player_score > cpu_score:
    print ("YOUR FINAL SCORE:", player_score)
    print ("COMPUTER FINAL SCORE:",cpu_score)
    print ("YOU WIN THE MATCH!")

if player_score == cpu_score:
    print ("YOUR FINAL SCORE:", player_score)
    print ("COMPUTER FINAL SCORE:",cpu_score)
    print ("IT'S A TIED MATCH!")
```
